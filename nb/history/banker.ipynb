```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.font_manager import FontProperties
# %matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 16)
import matplotlib
#matplotlib.style.use('ggplot')
matplotlib.style.use('default')
from numpy import *
import numpy as np

df=pd.read_table("./total",sep='\t',header=None,names=["time","err","adgroup","pv"])
c2a=pd.read_table("./c2a",sep='\t',header=None,names=["campaign","adgroup"])

l=7260
r=7270

for i in [0,204,205,206]:
    print "-----------------"+str(i)
    t=df[(df['err']==i)][['time','adgroup','pv']]
    m=pd.merge(t,c2a,left_on=['adgroup'],right_on=['adgroup'])
    k=m[(m['campaign']>=l)&(m['campaign']<=r)].groupby(['time','campaign'])['pv'].sum().reset_index(name='sum')
    pt=k.pivot('time','campaign','sum')
    try:
        
        pt.plot()
        plt.title("error:"+str(i)+"|"+str(l)+"-"+str(r))
    except Exception:
        pass
```
